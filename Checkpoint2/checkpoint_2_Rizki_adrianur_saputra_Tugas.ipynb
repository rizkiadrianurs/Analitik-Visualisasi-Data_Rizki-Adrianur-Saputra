### **Outliers Values**

```python
results = []
cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Berdasarkan hasil deteksi outliers, terlihat bahwa ada kolom yang mmemiliki persentase nilai ekstrem. Yaitu kolom sold memiliki persentase nilai ekstrem sebesar 13,6%, kolom available sebesar 10,1% dan price sebesar 5,4%. Hal ini menunjukkan adanya perbedaan nilai yang cukup jauh pada sebagian data, terutama pada jumlah penjualan dan ketersediaan produk. kondisi tersebut dapat memengaruhi hasil analisis dan model yang akan kita bangun nantinya, sehingga perlu tindakan lebih lanjut.


# **Eksplorasi Data (EDA)**

## **Comparison**

```python
penjualan_brand = df.groupby('brand')['sold'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan_brand.index,y=penjualan_brand.values,palette='viridis',hue=penjualan_brand.index,legend=False)

plt.title('10 Brand Teratas Berdasarkan Total Penjualan')
plt.xlabel('Brand Parfume')
plt.ylabel('Total Produk Terjual')
plt.xticks(rotation=45)
plt.show()
```

berdasarkan visualisasi terhadap 10 brand teratas, terlihat bahwa brand Calvin Klein dan Versace menjadi brand dengan total penjualan tertinggi dibandingkan brand lainnya. sementara itu, brand seperti davidoff, azzaro, dan armaf berada pada tingkat penjualan menengah, sedangkan brand lain memiliki total penjualan yang lebih kecil. hal ini menunjukkan bahwa konsumen cenderung lebih menyukai brand populer yang sudah memiliki reputasi kuat di pasar parfum.


## **Composition**

```python
sales_by_brand = df.groupby('type')['sold'].sum().sort_values(ascending=False)

sales_by_brand_top3 = sales_by_brand.head(3)

plt.figure(figsize=(10, 8))
sales_by_brand_top3.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)

plt.title('Proporsi 3 Type Teratas Berdasarkan Total Penjualan')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Berdasarkan diagram terhadap 3 type prafume terlaris, terlihat bahwa tipe eau de toilette mendominasi proporsi penjualan dengan persentase sekitar 86,8%, jauh lebih besar dibandingkan eau de parfum sebesar 8,4% dan eau de cologne sebesar 4,8%. hal ini menunjukkan bahwa konsumen lebih memilih eau de toilette, sehingga tipe ini menjadi type parfum yang paling banyak di beli oleh para konsumen.
